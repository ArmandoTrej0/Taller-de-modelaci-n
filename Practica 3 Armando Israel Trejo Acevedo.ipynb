## Bioacumulación
Las autoridades están preocupadas por los niveles elevados de contaminación con mercurio tóxico en los embalses que proveen el agua potable a la Ciudad de México, y solicitan asistencia para analizar la gravedad del problema. La ciencia ha estudiado los efectos adversos del mercurio en la salud humana por más de un siglo, y se sabe que las actividades humanas son las responsables por la mayor parte del mercurio en el medio ambiente. Por ejemplo, el mercurio, un subproducto del carbón, proviene de las emisiones de chimeneas de viejas centrales de carbón y se disemina por lluvia ácida. Sus partículas se elevan sobre las columnas de la chimenea y llegan a los vientos predominantes, que suelen soplar en dirección noreste. Después de colisionar con las montañas, las partículas caen a la tierra. Una vez en el ecosistema, los microorganismos en el suelo y los sedimentos del depósito destruyen el mercurio y producen un producto químico muy tóxico conocido como metilmercurio.

El mercurio experimenta un proceso llamado bioacumulación, que ocurre cuando los organismos absorben contaminantes más rápido que lo que tardan sus cuerpos en eliminarlos. Por lo tanto, la cantidad de mercurio en su sistema se acumula con el tiempo. Se sabe que los humanos pueden eliminar mercurio de su cuerpo a una tasa proporcional a la cantidad restante. Metilmercurio decae en un 50 % cada 65-75 días (vida media del mercurio) si no se ha ingerido más mercurio durante ese tiempo. En tu ciudad se han recogido y testeado 2425 muestras de lobina de los embalses, lo que produjo los siguientes datos: todos los peces estaban contaminados, y el valor medio de metilmercurio en la muestra de peces fue de 0.43 μ g (microgramos) por gramo. El peso promedio de los peces fue de 0.817 kg.

Tambien supondre que como el metilmercurio se reduce un 50% en 10 semanas,por semana este se reduce 5%  y mi unidad de tiempo seran semanas.

$Mc_{n}=$ Metilmercurio (en microgramos) en el cuerpo en la semana n

$Mc_{n+1}=Mc_{n}+(0.43*817)*2-(0.05*Mc_{n})$

```python
N=float(input('De que semana quieres ver la cantidad de metilmercurio en el cuerpo: '))
Mcn=0
j=0
Nt=[0]
Mcnt=[0]#Estos dos arreglos son para tener las cantidades semana a semana y poder graficar
while j<N:
  Mcn=Mcn+((0.43*817)*2)-(0.05*Mcn)
  Mcnt.append(Mcn)
  j=j+1
  Nt.append(j)
Mcnp=(Mcn*0.001)/70
print('La cantidad de Metilmercurio en cuerpo en la semana',N,'es de',Mcn,'microgramos, o',Mcnp,'miligramos por kilogramo')
#Este es mi codigo recursivo que dada la semana calcula el metilmercurio en el cuerpo con las supociciones que hice
```

```python
plt.plot(Nt,Mcnt)
plt.grid()
plt.ylabel('Metilmercurio en cuerpo (microgramos)')
plt.xlabel('Semanas')
#Observo que en las primeras 50 semanas la cantidad de metilmercurio se eleva de una manera muy rápida, aunque despues este llega a un equilibrio en un valor cercano a 14,000
```

## Modelo SI
Uno de los modelos compartimentales mas simples en epidemiología es el modelo SI (Susceptible-Infectado), presentado por Kermack y McKendrick en 1927. Dado una población de $N$ individuos, suponemos que dada una cantidad inicial $S_0$ de individuos susceptibles a una cierta enfermedad, y una cantidad $I_0$ de individuos infectados con dicha enfermedad, la evolución de la población (en días) durante una epidemia está gobernada por el siguiente sistema de ecuaciones en diferencia
$$
\begin{cases}
\Delta S_n=-\lfloor \beta S_nI_n \rfloor\\
\Delta I_n=\lfloor \beta S_nI_n \rfloor
\end{cases}
$$
 con $n\in \mathbb{N}$, dónde $\beta$ es la tasa de contagio de la enfermedad.
 - Comente sobre las principales hipotésis que debemos hacer sobre una enfermedad para poder aplicar este modelo. ¿Son razonables?
 - Utilizando la hipótesis que la población es constante ($S_n + I_n = N$ para todo $n ∈ \mathbb{N}$), escriba este sistema de ecuaciones en diferencia como una única ecuación en diferencia, de la que pueda deducir toda la información.
 - Comenzando de una población $(S_0,I_0) = (999,1)$ calcule y grafique (de manera bidimensional) el comportamiento de la población en los próximos 100 días si la tasa de contagio es β = 0,1.
 - El comportamiento a largo plazo, ¿depende de las condiciones iniciales?¿Cuáles son los distintos escenarios que podemos tener luego de mucho tiempo?

### A)

Las principales suposiciones que creo se hacen en este modelo son:

-Que la enfermedad no se cura, es decir, una ves infectado no puedes dejar de estarlo

-Toda la población es o susceptible o infectada, es decir no hay personas que no puedan infectarse

-La tasa de contagio es constante durante todo el tiempo que midamos



### B)

$S_{n}+I_{n}=N$

$S_{n}$ Es la cantidad de personas susceptibles el dia n

$I_{n}$ Es la cantidad de personas infectadas el dia n

$N$ Es la población total

$\Delta S_n=-\lfloor \beta S_n\frac{I_n}{N} \rfloor$

$\Delta I_n=\lfloor \beta S_n\frac{I_n}{N} \rfloor$

Hago este cambio de dividir entre N, pues investigue y esta Delta depende de la proporción de infectados respecto a la población total pues si usamos los valores de población directos esta delta toma valores muy grandes y no cobra sentido

$S_{n+1}=S_{n}+\Delta S_{n}=S_{n}-\lfloor \beta S_n\frac{I_n}{N} \rfloor$

$I_{n+1}=I_{n}+\Delta I_{n}=I_{n}+\lfloor \beta S_n\frac{I_n}{N} \rfloor$



Por lo que tomo la ecuación

$S_{n+1}=N-I_{n}-\lfloor \beta S_n\frac{I_n}{N} \rfloor$

y notemos que de ahi se puede deducir la otra pues

$S_{n+1}-N=-I_{n}-\lfloor \beta S_n\frac{I_n}{N} \rfloor$

$-I_{n+1}=-I_{n}-\lfloor \beta S_n\frac{I_n}{N} \rfloor$

$I_{n+1}=I_{n}+\lfloor \beta S_n\frac{I_n}{N} \rfloor$


### C)

```python
Inf=[999]
Sn=[1]
Di=[0]
De=[]
j3=0
b=0.1
while j3<99:
  sn1=1000-Inf[j3]-np.floor((b)*(Sn[j3]*(Inf[j3]/1000)))
  inf1=1000-sn1
  Delta=np.floor((b)*(Sn[j3]*(Inf[j3]/1000)))
  De.append(Delta)
  Inf.append(inf1)
  Sn.append(sn1)
  j3=j3+1
  Di.append(j3)
```

```python
plt.plot(Di,Inf,label='Infectados')
plt.legend(loc = 'lower right')
plt.plot(Di,Sn,label='Susceptibles')
plt.legend(loc = 'lower right')
plt.grid()
plt.xlabel('Dias')
plt.ylabel('Infectados-susceptibles')
```

Supongo que la proporcion se mantiene asi pues, al tomar $\Delta S_{n}=-\lfloor \beta S_n\frac{I_n}{N} \rfloor$ desde el principio $\frac{0.1}{1000} S_{0} I_{0}=0.0999$ y al tomar el piso de este valor, este es 0, por lo que los valores se quedan igual y esto se repite, por lo que la delta cambiaria si teniendo las mismas condiciones iniciales:

$\beta S_n\frac{I_n}{N}\geq 1$

$\beta \frac{999}{1000} \geq 1$

$\beta  \geq \frac{1000}{999}\approx1.001001001$

Y generalizando a diferentes datos tanto de poblacion como la proporcion inicial

$\beta  \geq \frac{N}{S_{0}I_{0}}$

Por lo que opino que el modelo no es muy preciso si la tasa de contagio no es tan alta, pues pienso que aunque esta se mantenga baja, igual influye en los contagios y en general el como se comporta la población entre infectados y susceptibles

## Modelo de supervivencia

Suponga que estamos estudiando la supervivencia de una cierta especie de ballenas, y sabemos que si el número de ballenas es menor que cierto nivel mínimo de supervivencia $m$, la especie se extingue. Suponga también que la población es limitada por una cierta población máxima $M$. Es decir, si la población de ballenas está por encima de $M$, ésta empezará a descender porque el ecosistema no puede soportar una población tan grande. Podemos modelar esta situación usando la siguente ecuación, donde an representa la población de ballenas luego de $n$ años. $$a_{n+1} −a_n = K(M −a_n)(a_n −m)$$ Discuta el modelo.

Al ser un modelo discreto podemos decir que

$a_{n+1}-a_{n}=\Delta a_{n}$

$\Delta a_{n}=K(M-a_{n})(a_{n}-m)$

Primero podemos saber por la hipotesis que $\Delta a_{n}$ es negativo cuando $a_{n}$ es mayor que $M$ o menor que $m$, y de ahi viene el multiplicar $(M-a_{n})(a_{n}-m)$, eso tambien quiere decir que $K>0$.

Podriamos ver a $K$ como la tasa de crecimiento poblacional.

Algo que puedo afirmar tambien es que si en algun momento $a_{n}$ toma el valor exacto de $m$ o $M$ este es un punto fijo en el modelo
