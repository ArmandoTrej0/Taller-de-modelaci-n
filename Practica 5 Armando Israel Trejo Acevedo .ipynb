## Ejercicio 1
El método del cuadrado nos permite obtener una sucesión de números aleatorios entre 0 y 9999. Este método fue desarrollado por J. Von Neuman, S. Ulm y N. Metropolis en 1964 en Los Alamos Laboratory, para simular colisiones de neutrones, como parte del proyecto Manhattan.  El gran problema de este método es que siempre degenera a cero, pero dependiendo de la semilla podría ser que la cantidad de pasos sea tan grande como queramos.

- Programe una función que, dada una semilla y un cierto máximo número $N$, devuelva una sucesión de números aleatorios en $[0,1]$ a lo más $N$ elementos.
- Si usamos como semilla $x_0=4450$  ¿Cuánto tiempo tardará en degenerar a cero? ¿Si usamos $x_0=4451$? ¿ Qué puedes conlcuir de esto?
- Muestre que este método aproxima la distribución uniforme.<br/>
Para esto, aproxime la función de distribución acumulada, dada por
	$$
	F_U(z)=P\left[\{\omega \in \Omega: U(\omega)\leq z\}\right], \qquad z\in \mathbb{R}
	$$
	usando frecuencias relativas de aparición.


### A)

```python
def metcuad (N,s,j=1): #La función te pedira la semilla, y la cantidad de números pseudoaleatorios que deseas
  S=[]#Fijo una lista donde agregare el número en cada iteración
  if 0<=s<=9999: #Pido que la semilla tenga a lo más 4 digitos y sea positiva
    while j<=N: #Mi ciclo while se rompera en el momento en el que ya hay N elementos en mi lista
      s=s**2 #Primero elevo al cuadrado mi elemento
      s=str(s) #Lo convierto en str para poder llamar cifra por cifra y elegir las de enmedio
      if len(s)==8:
        s=s[2]+s[3]+s[4]+s[5]
        s=int(s)
      elif len(s)==7:
        s=s[1]+s[2]+s[3]+s[4]
        s=int(s)
      elif len(s)==6:
        s=s[0]+s[1]+s[2]+s[3]
        s=int(s)
      elif len(s)==5:
        s=s[0]+s[1]+s[2]
        s=int(s)
      elif len(s)==4:
        s=s[0]+s[1]
        s=int(s)
      elif len(s)==3:
        s=s[0]
        s=int(s)
      elif len(s)<=2:
        s=int(0)
#Claramente no es la mejor opción, pero haciendo casos en base a la longitud de mi s^2, eligo los 6 números de enmedio como si pusiera los 0 faltantes para que la longitud de mi s^2 sea 8
      j=j+1 #añado un 1 a mi contador
      S.append(s) #El número pseudoaleatorio lo agrego a mi lista vacia
    return S
  else:
    print('Entrada de semilla no valida')
```

```python
C1=[] #Mi lista donde agregare números pseudoaletorios con mi función de metodos de cuadrados
for i in range (0,54): #el 54 es porque despues de ese número empieza un ciclo por lo que valores mayores ya no nos sirven
  C1.append((metcuad (54,4180)[i])/9999) #Hago esto para que los números esten entre [0,1]
  C1.sort() #Los acomodo de manera ascendente para no tener problemas al graficar más adelante, al ser equiprobable no hay mucho problema

C1s=[]
for i in range (1,55):
  C1s.append(i*(1/54))
#Como son pseudoaleatorios, podria decir que son equiprobables, por lo que la distribución acumulada del termino i seria i/54
```

## Ejercicio 2

Usando números pseudoaleatorios, aproxime el valor de la integral
$$
\int_{0}^{0.55}\tfrac{\Gamma(100)}{\Gamma(50)\Gamma(50)}x^{49}(1-x)^{49}dx,
$$
que corresponde a la probabilidad de que una variable aleatoria con distribución $Beta(50,50)$ tome valores menores que $0.55$.

```python
import random
```

```python
def F2(x):
  f=(np.math.factorial(99))/(np.math.factorial(49)*np.math.factorial(49))*(x**49)*((1-x)**49)
#Uso el factorial por la propiedad que tiene la función Gamma
  return f
#Hago mi función para poder citarla más adelante
```

```python
Dom=np.arange(0,0.5501,0.0001) #Hago una lista de 5501 elementos en el intervalo, para poder aplicarles la función y observar a detalle como se comporta
FDom=[F2(Dom[i]) for i in range (len(Dom))]
```

```python
plt.plot(Dom,FDom)
plt.grid()
#Al graficar podemos observar que esta acotada por el rectangulo de [0,0.55]*[0,8]
```

```python
AleatoX=[]
jx=0
#Lista donde agregare los elementos aleatorios en mi intervalo de X, es decir [0,0.55]
while jx<=5500:
  r2x=random.random()*0.55
  AleatoX.append(r2x)
  jx=jx+1
#Multiplico pues random.random da elementos entre 0 y 1, y con esa multiplicación da elementos entre 0 y 0.55

AleatoY=[]
jy=0
while jy<=5500:
  r2y=random.random()*8
  AleatoY.append(r2y)
  jy=jy+1
#Hago lo mismo que antes solo que multiplico por 8, pues busco elementos entre 0 y 8
```

```python
plt.plot(Dom,AleatoY,'.')
```

```python
#Ahora se que el Area del rectangulo entre el area por debajo de la función tiene una proporción, calculare un aproximado de mi proporción usando los puntos pseudoaleatorios
Cont=[0]*len(AleatoY) #Hago una lista de elementos todos 0 de longitud 5501, donde checare 1 por 1 si el punto (X,Y) es menor o no que el punto(X,f(X))
for i in range (len(Cont)):
  if AleatoY[i]<=F2(AleatoX[i]): #Checara si el elemento de Y asociado al de X es menor que la función en X, y si eso pasa, su lugar en la lista Cont cambiara a 1
    Cont[i]=1
  else:
    Cont[i]=0
#Para saber la cantidad de puntos que lo cumplen solo sumare los elementos de Cont
sum(Cont)
```

Ahora ya tengo aproximadamente la proporción, entonces podria afirmar

$\tfrac{A_{Función}}{A_{Rectangulo}}\approx \tfrac{1056}{5501}$

Por lo que solo falta calcular el area del rectangulo, pero es sencilla pues lo definimos como el rectangulo de los intervalos $[0,0.55]*[0,8]$ por lo que su area seria $(0.55-0)(8-0)=0.55*8=4.4$, por lo que ahora se:

$\tfrac{A_{Función}}{4.4}\approx \tfrac{1056}{5501} ⟹ A_{Función}\approx\tfrac{4*1056}{5501}=0.76786038902$

**Dado Cargado**:<br>
Ahora, queremos modelar un dado cargado con $k$ caras; es decir que no todas las caras tienen la misma probabilidad de salir. Para poder simular este dado su función deberá recibir un vector de probabilidades $(p_1 ,p_2 ,...,p_k)$ donde $p_i$ es la probabilidad de obtener la i-ésima cara del dado. Recuerden que se debe cumplir que $p_1 +p_2 +...+p_k =1$. <br>
**Pista**: Su función debe ser muy similar a la del dado justo pero ahora en lugar de aparecer las probabilidades $1/k$ debarán aparecer las $p_i$.

### Dado justo

```python
def dadojusto(k):
  k=int(k)
  x=random.random()*k
  if x==k:
    y=x
  else:
    y=int(x)+1
  print(f'Salio la cara {y}')
```

```python
dadojusto(10)
```

### Dado cargado

```python
def dadocargado (k):
  Probabilidades=[0]*k #Primero hago una lista de rango k para almacenar las probabilidades
  Probabilidades[0]=random.random() #Hago la primera probabilidad con un random
  for i in range (len(Probabilidades)-2): #Para las demas probabilidades (Exceptuando la ultima)
    Probabilidades[i+1]=random.random()*(1-sum(Probabilidades)) # pido un random que tenga como valor maximo lo que queda para completar el 1 sumar las probabilidades ya obtenidas
  Probabilidades[-1]=1-sum(Probabilidades)# Para la ultima probabilidad solo agrego lo que falte para completar el 1 si sumo las demas probabilidades
  x=random.random() #Hago el tiro que me de un número random entre 0 y 1
  for i in range (len(Probabilidades)): #Revisare mi "tiro" en comparación a los elementos de mi lista de probabilidades
    A=Probabilidades[:i]
    B=Probabilidades[:i+1]
    if x==1:
      Y=k#Esto es solo para el caso en donde x=1, pues en ese caso con el resto del codigo me apareceria que la cara es la k+1 y no tiene sentido
      break
  #Defino a A y B como las listas de Probabilidades hasta el elemento i y el elemento i+1 (sin contarlos) respectivamente (Aqui no afecta que i+1 le gane a mi lista pues si es mayor solo lo ignora)
    elif sum(A)<x<=sum(B): #Si se cumple que mi random x, esta entre la suma de probabilidades hasta el indice i y la suma hasta el indice i+1, significa que es la cara i+1
      Y=i+1 #Fijo mi cara con la variable Y (es i+1 pues el conteo inicia desde i=0, pero eso significaria que salio la cara 1)
      break #Si se cumple esto, termino mi ciclo for para que no se haga otra iteracióm

  print(f'Salio la cara {Y}') #Por ultimo imprimo la cara que salio
  print('La lista de probabilidades fue:',Probabilidades)
  print('El random de tiro fue',x) #Esto es solo para checar que el codigo funcióna
```

```python
dadocargado(6)
#Es facil observar que el dado al estar cargado y por como realice mi repartición, se puede ver que las caras más grandes casi siempre les tocara menos probabilidad de aparición, pues dependen de que a todas las menores que esta le toque una probabilidad baja
```
