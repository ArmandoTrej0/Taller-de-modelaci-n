## Ejercicio 2:
Haz una función que simule un estado dado una distribución de una cadena de Markov considerando una distribución inicial. $$v=(x_1 ,x_2 ,...,x_n)$$ Si numeramos los estados de la cadena comenzando con 0 la idea es exactamente la misma que cuando programaron el dado cargado en la práctica 10. Su función debe recibir:
- Un vector de probabilidades cuya suma sea 1

Debe regresar de manera aleatoria (con la probabilidad de elección correspondiente) una de las etiquetas de los estados.

```python
def Elección (Vprob): #Pido el vector de probabilidad
  X=[1]*len(Vprob)
  Y=[]
  for i in range (len(X)):
    Y.append(sum(X[:i+1]))
  #Esto es para hacer una lista Y que me indique la "Cara" asociada a cada probabilidad
  x=random.random()#Elijo un número random entre 0 y 1
  Probs=[]
  for i in range (-1,len(Vprob)):
    Probs.append(sum(Vprob[:i+1]))
    #Hago mi lista "Probs" con las probabilidades acumuladas para poder asociar mi random a cada probabilidad
  for i in range (len(Probs)-1):
    if Probs[i]<x<=Probs[i+1]: #Si se cumple que mi random esta entre la probabilidad acumulada i y la i+1 (Como la lista Probs inicia en 0), se tomara como efectiva la probabilidad i+1
      return (Y[i])
        #Como Probs inicia en 0, tiene un elemento de más, por lo que el Probs[i+1] es la acumulada de Vprob[i], entonces elijo la cara i de mi lista Y
Elección([1/4,1/4,1/4,1/4])
```

I**Idea para simular una cadena de Markov**<br>
Necesitamos generar una sucesión $\{x_0 ,x_1 ,x_2 ,...,x_N\}$ dado un espacio de estados $E$.
Supongamos un espacio de estados $$E=\{0,1,2\}, π=\left(\frac{1}{3},\frac{1}{3},\frac{1}{3}\right),
P=\begin{bmatrix}
\frac{1}{2} & \frac{1}{2} & 0 \\
\frac{1}{2} & 0 & \frac{1}{2} \\
0 & \frac{1}{2} & \frac{1}{2}
\end{bmatrix}
$$
1. Simulamos $x_0$. Para esto necesitamos el vector de distribución inicial $π=\left(\frac{1}{3},\frac{1}{3},\frac{1}{3}\right)$.

Escogemos un estado aleatorio con la función anterior y $π$. Listo! tenemos $x_0$ supongamos que $x_0 =2$.

2. Simulamos $x_1$. Una cadena de Markov depende del estado en que estuve antes, entonces tenemos que utilizar la matriz de trasición y fijarnos en el renglón del estado en que estabamos.
$$P[2,:]= \left(0,\frac{1}{2},\frac{1}{2}\right)$$

Escogemos un estado aleatorio con la función anterior y P[2,:].
Listo! Repetimos... hasta n.

## Ejercicio 3:
En este ejercicio programarán dos funciones que simule una cadena de Markov y los estados que va tomando en cada iteración. La diferencia entre estas funciones será el criterio de paro.
- La primera función deberá hacer: :
a. N iteraciones de la cadena de markov y la deberá recibir como parámetro.
b. Regresar la lista de los estados que recorrió.
- Mientras que la segunda función:
1. deberá correr hasta que la cadena tome un estado e dado. Deberá recibir la etiqueta del estado al que se quiere llegar, y un número máximo de iteraciones, MaxIter , en caso de que nunca se alcance el estado deseado.

Ambas funciones deben recibir **dos parámetros correspondientes a la cadena**.
- El primer parámetro es deberá ser la matriz de transición de la cadena, $P$.
- El segundo parámetro deberá ser la distribución inicial de la cadena, $\Pi_0$.

```python
def cadenaMarkovN (Matr,Vect,N):#Pedire la matriz de transición Matr, mi vector de estado inicial Vect y N numero de iteraciónes
  #Estoy suponiendo que mi espacio de Estados es como un dado, es decir los naturales 1 a n donde n es el tamaño de mi Espacio de Estados
  I=Elección(Vect) #Elijo mi Estado inicial
  Pasos=[I] #La lista donde agregare los valores en cada transición
  while len(Pasos)<N: #Hare que se repita mientras su longitud sea menor a N (la ultima iteración hara que su longitud sea N)
    Z=Matr[I-1] #Una ves mi estado, elijo su fila en la matriz de transicón (Es la I-1 porque las listas empiezan su conteo en 0 y mi I empieza siendo minimo 1)
    I=Elección(Z) #Nuevamente elijo un nuevo I con mi función eleccion en la fila de transición correspondiente
    Pasos.append(I) #Agrego el nuevo paso a mi lista Pasos
  return Pasos #Regreso la lista Pasos
```

```python
#Usaremos la matriz P1 del ejercicio 1 y el vector de estado inicial sera [1/3,1/3,1/3]
V3=[1/3,1/3,1/3]
TR=cadenaMarkovN(P1,V3,50)
print(TR)
```

```python
def cadenaMarkovEst(Matr,Vect,N,x):
  I=Elección(Vect)
  Pasos=[I]
  while len(Pasos)<N and Pasos[-1]!=x:
    #El codigo es muy parecido solo que el while se ejecutara mientras la longitud de mi lista pasos sea menor a N y su ultimo elemento sea diferente al x que queremos
    Z=Matr[I-1]
    I=Elección(Z)
    Pasos.append(I)
  return print(Pasos)
```

```python
TR2=cadenaMarkovEst(P1,V3,50,3)
TR2
```
